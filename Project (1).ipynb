```python
pool_list = []

for file in filenames:
    h = open(file, 'r') 

    # Reading from the file 
    content = h.readlines()

    with open(file) as f:
        for line in f:
            pass
        grade = line.strip()

    temp = []

    for line in content[:-2]:
        for i in line:  
        # Checking for the digit in the string
            if i.isdigit() == True:
                copies = int(i)

        words = line[2:].strip()
        for i in range(copies):
            temp.append(card_dict[words])

    listy = []
    listy.append(grade)
    for card in setnums:
        if card in temp:
            listy.append(temp.count(card))
        else:
            listy.append(0)
    
    pool_list.append(listy)
    
```

```python
def parse(f):
	x = []
	y = []
	line_num = 0
	for line in open(f):
		if(line_num > 0):
			split = line.split(",")
			x_feats = []
			for i in range(len(split)):
				if(i == 1):
					y.append(split[i])
				if(i >= 2):
					if(i == len(split) - 1):
						x_feats.append(int(split[i][:len(split[i])]))
					else:
						x_feats.append(int(split[i]))
			x.append(x_feats)

		line_num += 1
	return (x,y)


(x,y_unclean) = parse("draft_pools_2")
```

```python
grade_dict = {
    'A+':0,
    'A':1,
    'A-':2,
    'B+':3,
    'B':4,
    'B-':5,
    'C+':6,
    'C':7,
    'C-':8,
    'D+':9,
    'D':10,
    'D-':11,
    'F+':12,
    'F':13
}

def label_encode(L):
	global grade_dict
	newL = []
	for i in range(len(L)):
		newL.append(grade_dict[L[i]])
	return newL
y = label_encode(y_unclean)

combined = copy.deepcopy(x)
for sample in range(len(combined)):
	combined[sample].append(y[sample])
 
columns_df = []

for i in range(len(combined[0])):
	if(i != len(combined[0])-1):
		columns_df.append('card'+str(i))
	else:
		columns_df.append('grade')

# print(len(combined[0]))
data = combined

####### END PARSING #############

# columns: card0, card1, ... , card285, grade
df = pd.DataFrame(data, columns = columns_df)
# print(df)

x = df[columns_df[:len(columns_df)-1]]
y = df['grade']
```

### Neural Network

```python
layers = []
for layer_size in [100,200,500]:
	for num_layers in [2,5,10]:
		size_list = [layer_size for _ in range(num_layers)]
		size_tuple = tuple(size_list)
		layers.append(size_tuple)

best_model = None
best_diff = 999999
for net in layers:
	print("HIDDEN LAYERS: " + str(net))
	mlp = MLPClassifier(hidden_layer_sizes=net, activation='relu', solver='sgd', max_iter=5000)
	mlp_model = mlp.fit(X_train,Y_train)
	predict_train = mlp_model.predict(X_train)
	predict_test = mlp_model.predict(X_test)
# 	print(predict_test)
# 	print("train acc = " + str(accuracy_score(predict_train,Y_train)))
# 	print("test acc = " + str(accuracy_score(predict_test,Y_test)))
# 	print("f1 score train = " + str(f1_score(predict_train,Y_train,average='weighted')))
# 	print("f1 score test = " + str(f1_score(predict_test,Y_test,average='weighted')))
# 	print("rmse train = " + str(math.sqrt(mean_squared_error(predict_train,Y_train))))
# 	print("rmse test = " + str(math.sqrt(mean_squared_error(predict_test,Y_test))))
	mse_train = math.sqrt(mean_squared_error(predict_train,Y_train))
	mse_test = math.sqrt(mean_squared_error(predict_test,Y_test))
	if(abs(mse_test-mse_train) < best_diff):
		best_model = mlp_model
		best_diff = abs(mse_test-mse_train)
    
print(metrics.classification_report(Y_train, predict_train))
print(metrics.classification_report(Y_test, predict_test))
```
